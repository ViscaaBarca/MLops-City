```python
# Imports
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, trim, regexp_replace, to_timestamp, year, month, dayofmonth, hour, dayofweek
from pyspark.sql.types import StringType

spark = SparkSession.builder.getOrCreate()

# Pad
train_path = '/Workspace/Users/20195060@student.hhs.nl/MLops-City/Metro_Interstate_Traffic_Volume_train.csv'
test_path  = "/Workspace/Users/20195060@student.hhs.nl/MLops-City/Metro_Interstate_Traffic_Volume_test.csv"
# functie voor schoonmaken
def clean_df(df):

    # 1. duplicaten verwijderen
    df = df.dropDuplicates()

    # 2. datum omzetten
    df = df.withColumn("date_time", to_timestamp(col("date_time"), "yyyy-MM-dd HH:mm:ss"))

    # 3. string kolommen opschonen
    string_cols = [f.name for f in df.schema.fields if isinstance(f.dataType, StringType)]
    for c in string_cols:
        df = df.withColumn(c, trim(col(c)))
        df = df.withColumn(c, regexp_replace(col(c), r"\s+", " "))

    # 4. numerieke kolommen bepalen
    numeric_types = ["int", "bigint", "double", "float", "long", "short", "decimal"]
    numeric_cols = [f.name for f in df.schema.fields if f.dataType.simpleString() in numeric_types]

    # 5. missende numerieke waardes vullen (mediaan)
    for c in numeric_cols:
        median_value = df.approxQuantile(c, [0.5], 0.01)[0]
        df = df.fillna({c: median_value})

    # 6. missende string waardes vullen
    for c in string_cols:
        df = df.fillna({c: "Unknown"})

    # 7. datum features maken
    df = (
        df
        .withColumn("year", year(col("date_time")))
        .withColumn("month", month(col("date_time")))
        .withColumn("day", dayofmonth(col("date_time")))
        .withColumn("hour", hour(col("date_time")))
        .withColumn("dayofweek", dayofweek(col("date_time")))
    )

    return df


# Alles in 1 lijn uitvoeren (dan wordt neiwue df een keer opgeslagen)
df_train = clean_df(
    spark.read.option("header", True).option("inferSchema", True).csv(train_path)
)

df_test = clean_df(
    spark.read.option("header", True).option("inferSchema", True).csv(test_path)
)

# Resultaat bekijken
display(df_train.limit(5))
display(df_test.limit(5))
```

## Edge Pipeline: YOLOv8 Training & Evaluatie

**Onderbouwing Modelkeuze (YOLOv8n):**
Voor de slimme camera's op de kruispunten (Edge devices zoals een Raspberry Pi of NVIDIA Jetson) is rekenkracht beperkt. We hebben gekozen voor de *nano* variant (`yolov8n.pt`) omdat deze een modelgrootte heeft van slechts ~6 MB en hoge FPS (Frames Per Second) garandeert tijdens inferentie. Zwaardere varianten (zoals YOLOv8m of YOLOv8x) zijn trager en verhogen de latency, wat ongewenst is voor realtime adaptieve verkeerslichten.

**Onderbouwing Optimizer (AdamW vs SGD):**
Standaard gebruikt YOLO SGD (Stochastic Gradient Descent). Wij hebben specifiek gekozen voor **AdamW** via hyperparametertuning. Bij relatief beperkte datasets en transfer learning (we finetunen een pre-trained model) convergeert AdamW vaak sneller en stabieler zonder dat we een uiterst complexe learning-rate schedule hoeven in te stellen, zoals bij SGD vereist is.

**Modelvergelijking / Tuning opmerking:**
In eerdere MLOps-iteraties hebben we geëxperimenteerd met YOLOv5 en verschillende epochs. De gekozen configuratie (50 epochs, AdamW) bleek de optimale balans tussen mAP score en trainingsduur te bieden ter voorkoming van overfitting.

```python
mlflow.set_experiment("/Shared/YOLOv8_Traffic_Detection")

with mlflow.start_run(run_name="yolov8n_edge_model"):
        
    actual_epochs = 50
    actual_batch = 16
    actual_optimizer = "AdamW"
    
    mlflow.log_param("model", "yolov8n")
    mlflow.log_param("epochs", actual_epochs)    
    mlflow.log_param("imgsz", 640)
    mlflow.log_param("batch", actual_batch)    
    mlflow.log_param("optimizer", actual_optimizer)
    mlflow.log_param("edge_target", "Jetson Nano")

    # YOLO model laden (Transfer learning vanaf pretrained weights)
    model = YOLO("yolov8n.pt")

    print("Start YOLO training. (Zie logs voor voortgang)")
    start_time = time.time()

    # Training starten 
    yolo_results = model.train(
        data="data.yaml",
        epochs=actual_epochs,
        imgsz=640,
        batch=actual_batch,
        optimizer=actual_optimizer,
        amp=True,
        patience=10, 
        cache=True,
        workers=4,
        project="traffic_yolo",
        name="yolov8n_edge"
    )

    training_time = time.time() - start_time
    print(f"Training voltooid in {training_time:.2f} seconden")

    # Metrics ophalen uit de resultaten
    metrics = model.val()

    # Metrics netjes loggen in MLflow
    mlflow.log_metric("training_time_seconds", training_time)
    mlflow.log_metric("mAP50", float(metrics.box.map50))
    mlflow.log_metric("mAP50_95", float(metrics.box.map))
    mlflow.log_metric("precision", float(metrics.box.mp))
    mlflow.log_metric("recall", float(metrics.box.mr))

    # Modelgrootte bepalen (Belangrijk voor Edge Deployment!)
    best_model_path = "traffic_yolo/yolov8n_edge/weights/best.pt"
    if os.path.exists(best_model_path):
        size_mb = os.path.getsize(best_model_path) / (1024 * 1024)
        mlflow.log_metric("model_size_mb", size_mb)
        
        # Log het fysieke model naar MLflow
        mlflow.log_artifact(best_model_path, artifact_path="yolo_weights")

    print("YOLO metrics en model succesvol gelogd naar MLflow!")
```

## Edge Pipeline: Model Optimalisatie & TFLite Export

**Voorbereiding voor Edge Deployment**
Hoewel het zojuist getrainde YOLOv8n model in PyTorch-formaat (`.pt`) al relatief klein is, is dit framework niet optimaal voor snelle inferentie op low-power edge-devices (zoals de processor van een slimme camera of een Raspberry Pi).

Om te garanderen dat we voldoen aan de harde eisen uit ons Productvereistendocument (PRD) — *Modelgrootte < 10 MB* en *Inferentiesnelheid < 50 ms* — exporteren we het model naar **TensorFlow Lite (.tflite)**. 

Hierbij maken we gebruik van **INT8 Quantization** (`int8=True`). Dit compressieproces zet de interne wiskundige gewichten van het model om van zware 32-bit getallen naar lichte 8-bit getallen. Dit levert twee grote voordelen op voor ons MLOps-systeem:
1. **Grootte:** Het model wordt gehalveerd qua opslagruimte (vaak naar slechts ~3 MB).
2. **Snelheid:** Inferentie op kleine edge-CPU's wordt aanzienlijk sneller, met een verwaarloosbaar verlies in accuraatheid (mAP).

```python
# Laad het zojuist getrainde beste model in
best_model = YOLO("traffic_yolo/yolov8n_edge/weights/best.pt")

# Exporteer naar TFLite
export_path = best_model.export(format="tflite", int8=True)

print(f"TFLite export succesvol! Opgeslagen in: {export_path}")

# Log het TFLite model ook in MLflow zodat je het nooit kwijtraakt
if os.path.exists(export_path):
    mlflow.log_artifact(export_path, artifact_path="yolo_tflite_edge_model")
    
    # Bereken hoe klein hij is geworden en log dit
    tflite_size_mb = os.path.getsize(export_path) / (1024 * 1024)
    mlflow.log_metric("tflite_model_size_mb", tflite_size_mb)
    print(f"Het TFLite model is slechts {tflite_size_mb:.2f} MB groot!")
```
